```python
import os
import unicodedata

# Caminho da pasta onde estão as imagens
pasta_imagens = '.'  # ou use: os.getcwd() se preferir

def formatar_nome(nome):
    nome_sem_extensao, extensao = os.path.splitext(nome)
    nome_formatado = unicodedata.normalize('NFD', nome_sem_extensao)
    nome_formatado = nome_formatado.encode('ascii', 'ignore').decode('utf-8')
    nome_formatado = nome_formatado.replace(" ", "_")
    nome_formatado = ''.join(c for c in nome_formatado if c.isalnum() or c in ['_', '-'])
    nome_formatado = nome_formatado.lower()
    return nome_formatado + extensao.lower()

def mostrar_previa_renomeacao(pasta):
    print("Prévia da renomeação de arquivos:")
    print("-" * 50)
    for nome_original in os.listdir(pasta):
        caminho_antigo = os.path.join(pasta, nome_original)
        if os.path.isfile(caminho_antigo):
            novo_nome = formatar_nome(nome_original)
            if nome_original != novo_nome:
                print(f'{nome_original}  →  {novo_nome}')
    print("-" * 50)
    print("Nenhuma alteração foi feita ainda.")

def renomear_arquivos(pasta):
    print("Renomeando arquivos...")
    count = 0
    for nome_original in os.listdir(pasta):
        caminho_antigo = os.path.join(pasta, nome_original)
        if os.path.isfile(caminho_antigo):
            novo_nome = formatar_nome(nome_original)
            caminho_novo = os.path.join(pasta, novo_nome)
            if nome_original != novo_nome and not os.path.exists(caminho_novo):
                os.rename(caminho_antigo, caminho_novo)
                print(f'Renomeado: {nome_original} → {novo_nome}')
                count += 1
            else:
                print(f'⚠️  Pulado (já existe): {novo_nome}')

    print(f"\nTotal de arquivos renomeados: {count}")

# Passo 1: Mostrar prévia
mostrar_previa_renomeacao(pasta_imagens)

# Passo 2: Renomear (descomente a linha abaixo quando estiver pronto)
renomear_arquivos(pasta_imagens)
```
